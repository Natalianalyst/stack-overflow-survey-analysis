Завдання 2. Аналіз повноти відповідей респондентів

```python
schema_questions = set(schema['qname'])
df_columns = set(df.columns)
```

```python
# Знаходимо спільні колонки
common_questions = schema_questions & df_columns

# Додаємо ResponseId вручну
common_questions.add('ResponseId')

# В список
common_questions = list(common_questions)

# Фільтруємо датасет
df_filtered = df[common_questions]

# Видаляємо рядки з пропусками
complete_answers = df_filtered.dropna()

# Рахуємо унікальних респондентів
complete_count = complete_answers['ResponseId'].nunique()

print("Кількість респондентів без пропусків:", complete_count)
```

Завдання 3. Статистичний аналіз досвіду респондентів

```python
work_exp = df['WorkExp'].dropna()

mean_exp = work_exp.mean()
median_exp = work_exp.median()
mode_exp = work_exp.mode()[0]

stats = pd.Series({
    "Mean": mean_exp,
    "Median": median_exp,
    "Mode": mode_exp})
stats
```

Середній досвід респондентів становить приблизно 13 років, медіанний та модальний — 10 років, що свідчить про концентрацію більшості відповідей навколо цього значення.

Завдання 7. Географічний аналіз компенсації Python-розробників

```python
df[['Country', 'ConvertedCompYearly']].head()
```

```python
python_devs = df[df['UsesPython']]

# Забираємо рядки без зарплати і країни
python_devs = python_devs.dropna(subset=['ConvertedCompYearly', 'Country'])

# Групуємо по країнах і рахуємо статистику
salary_by_country = (python_devs.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).reset_index())

# Перейменуємо колонки
salary_by_country.columns = ['Country', 'AverageCompensation', 'MedianCompensation']

# Сортуємо
salary_by_country = salary_by_country.sort_values(by='AverageCompensation', ascending=False)

salary_by_country.head(10)
```

```python
count_by_country = python_devs.groupby('Country')['ResponseId'].nunique()

salary_by_country['Count'] = count_by_country.values

salary_by_country.head(10)
```

Аналіз компенсації Python-розробників у розрізі країн показав, що найвищі середні показники мають країни з дуже малою кількістю респондентів (наприклад, Оман, Андорра, Швейцарія). Це свідчить про те, що на середнє значення в таких випадках суттєво впливає високий дохід окремих респондентів.

У країнах з більшою вибіркою (наприклад, Нігерія, Ірландія, Сінгапур) спостерігається значна різниця між середнім значенням та медіаною, що свідчить про нерівномірний розподіл доходів. У таких випадках медіана є більш надійним показником.

Тому при аналізі компенсації Python-розробників доцільно враховувати як статистичні показники, так і кількість учасників опитування.

Завдання 9. Аналіз популярності Python по віковим категоріям

```python
df['Age'].value_counts().head(10)
```

```python
age_df = df.dropna(subset=['Age'])

python_by_age = (age_df.groupby('Age').agg(total=('ResponseId', 'nunique'), python_users=('UsesPython', 'sum')).reset_index())

python_by_age['PythonPercent'] = (python_by_age['python_users'] / python_by_age['total']) * 100

python_by_age.sort_values('Age')
```

Аналіз популярності Python за віковими категоріями показав, що найвищий відсоток користувачів Python спостерігається серед респондентів віком 18–24 роки (близько 40%). Це свідчить про високу популярність Python серед початківців та молодих спеціалістів.

У вікових групах від 25 до 54 років частка Python-розробників залишається стабільною (приблизно 36–38%), що вказує на довгострокову актуальність мови у професійному середовищі.

У групі 65+ років відсоток дещо знижується, що може бути пов’язано з меншою готовністю до освоєння нових технологій.

Загалом, Python є популярною мовою програмування у всіх вікових групах, особливо серед молодих розробників.
